```python
import random

coin = random.choice(["heads", "tails"])
print(coin)

```

```python
from random import choice

coin = choice(["heads", "tails"])
print(coin)

```

```python
import sys
# after module name everything in () is parameter and this can be
# used in multiple ways
if len(sys.argv) <2:
    print("Too few arguments")
elif len(sys.argv)>2:
    print("Too many arguments")
else:
    print('Hello, my name is ', sys.argv[1])

'''
You can pass input from terminal as well :
python3 Test.py DAVID
Hello, my name is  DAVID

if you enter o as sys.argv parameter it will print file name
'''
```

```python
import sys
# after module name everything in () is parameter and this can be
# used in multiple ways

if len(sys.argv) <2:
    sys.exit("Too few arguments")
for arg in sys.argv[1:]:
    print('Hello, my name is ', arg)

'''
Use sys.exit to exit process if it doesn't meet requirements
Everything you put in while running code from terminal it will accept it as
new item to print

> python3 Test.py david tex ben
Hello, my name is  david
Hello, my name is  tex
Hello, my name is  ben
> python3 Test.py david ,  tex ,  ben
Hello, my name is  david
Hello, my name is  ,
Hello, my name is  tex
Hello, my name is  ,
Hello, my name is  ben

'''
```
