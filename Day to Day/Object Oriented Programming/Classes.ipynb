```python
# This is a more complete implementation of the Student class, it includes validation for the name and house, it also includes error handling for the get_students function, it raises a ValueError if the name or house is missing or if the house is not valid. This way we can ensure that the data we are storing in our class is valid and we can also handle errors gracefully.
class Student:
    def __init__(self, name: str, age: int, email: str):
        name = name.strip()
        email = email.strip()

        if not name:
            raise ValueError("Name cannot be empty")

        if age < 5 or age > 120:
            raise ValueError("Age must be between 5 and 120")

        if "@" not in email or "." not in email:
            raise ValueError("Email looks invalid")

        self.name = name
        self.age = age
        self.email = email


def main():
    name = input("Enter name: ")
    age = int(input("Enter age: "))
    email = input("Enter email: ")
    
    try:
        s = Student(name,age,email )
        print(f"Student is registered with Name: {s.name} Age: ({s.age})and Email -> {s.email}")
    except ValueError as e:
        # here whatever the error message was in the ValueError, 
        # it will be captured in variable e and we can print it out or log it
        print("Could not create student:", e)


if __name__ == "__main__":
    main()
```
