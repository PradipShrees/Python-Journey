```python
# @classmethpd
"""
	•	No class indentation → function
	•	Inside class + self → instance method
	•	Inside class + @classmethod + cls → class method
	•	Inside class + @staticmethod → static method
	•	something.method() → method from type(something) (object’s class) or the class itself
"""
import random
class Hat:
   
    houses = ["Gryffindor", "Hufflepuff", "Ravenclaw", "Slytherin"]
    # houses is a class variable that contains a list of the four Hogwarts houses. This variable is shared among all instances of the Hat class, meaning that it can be accessed and modified by any instance of the class. In this case, it is used to store the names of the houses that students can be sorted into.
    @classmethod
    def sort(cls, name):
      #here cls.name is a local variable that holds the name passed to the sort method. The random.choice function is used to select a random house from the cls.houses list, and the result is printed in the format "name is in house".
      # and def sort is a method that takes a name as an argument and randomly assigns that name to one of the houses in the cls.houses list, printing the result. 
        print(name,"is in", random.choice(cls.houses))

Hat.sort("Harry")
Hat.sort("Ron")
Hat.sort("Hermione")
# here Hat.sort is a call to the sort method of the Hat class, passing in the names "Harry", "Ron", and "Hermione" as arguments. Each time the sort method is called, it will randomly assign the name to one of the houses in the cls.houses list and print the result.

# The code defines a class called Hat,
# which has a class variable called houses that contains a list of the four Hogwarts houses. The class also has a class method called sort, which takes a name as an argument and randomly assigns that name to one of the houses in the houses list, printing the result. Finally, the code calls the sort method three times with different names to demonstrate how it works.
# The output will be different each time you run the code, as the house assignment is random.
```

```python
# @ Static methods are methods that belong to the class rather than an instance of the class.
#  They can be called on the class itself, without needing to create an instance of the class. 
# Static methods are defined using the @staticmethod decorator.

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @staticmethod
    def is_adult(age):
        return age >= 18
    
# Example usage:
print(Student.is_adult(20))  # Output: True
print(Student.is_adult(15))  # Output: False
# In this example, the is_adult method is a static method that checks if a given age is 18 or older. It can be called directly on the Student class without needing to create an instance of the class.
# Note: staticmethod can return any value, and it does not have access to the instance (self) or class (cls) variables. It is simply a function that is associated with the class.
```
