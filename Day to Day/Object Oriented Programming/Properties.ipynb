```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Student name is {self.name} and age is {self.age}"
    
    @classmethod
    def get(cls):
        name = input("Enter the name of the student: ")
        age = int(input("Enter the age of the student: "))
        return cls(name, age)
def main():
    student = Student.get()
    print(student)

if __name__ == "__main__":
    main()
# nOTE: main() is called at last so it doesn't matter where we define the class and the main function. We can define the main function before the class as well. The main function is just a convention to indicate that this is the entry point of the program.

"""
	•	No class indentation → function
	•	Inside class + self → instance method
	•	Inside class + @classmethod + cls → class method
	•	Inside class + @staticmethod → static method
	•	something.method() → method from type(something) (object’s class) or the class itself
    •	something.attribute → attribute from type(something) (object’s class) or the class itself
"""
```

```python
class Amount:
    def __init__(self):
        self._amount = 0 # _amount is a private variable to store the balance

    @property # This decorator allows us to access the balance as a property instead of a method
    def balance(self):
        return self._amount
    
    def deposit(self, n):
        self._amount += n

    def withdraw(self, n):
        self._amount -= n

def main():
    amount = Amount()
    print(f"Initial Balance: {amount.balance}")  # Output: 0
    amount.deposit(100)
    print(f"Balance after deposit: {amount.balance}")  # Output: 100
    amount.withdraw(50)
    print(f"Balance after withdrawal: {amount.balance}") 
if __name__ == "__main__":
  main() # Output: 50
```
