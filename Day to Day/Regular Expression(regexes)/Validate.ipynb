```python
email = input("Enter your email: ").strip()
#strip removes any leading or trailing whitespace from the input

if "@" in email and "." in email:
    print("Valid")
else:
    print("Invalid")
#This code checks if the email 
# contains both "@" and "." 
# characters, which are typically required for a
#  valid email address. If both characters are 
# present, it prints "Valid"; otherwise,
#  it prints "Invalid".
```

```python
email = input("Enter your email: ").strip()
#strip removes any leading or trailing whitespace from the input

username, domain =email.split("@")

if (username) and ("." in domain):
    print("Valid")
else:
    print("Invalid")


"""This is another variation of the code that 
uses regular expressions to validate the
 email address.
 
 note how split is used to separate the username and 
 domain, and then a regular expression is used to 
 check if the email address is valid."""
```
