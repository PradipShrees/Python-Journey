```python
def total(galleons, sickles, knuts): # if you add default value of param = 0, it will make optional to pass all value in dictionary.
    return(galleons * 17 + sickles) * 29 + knuts
"""
coins = (100,50,25)
print(total(coins[0], coins[1], coins[2]), "Knuts") 
this works but for bigger data it's unsustainable
"""
coins = {"galleons": 100, "sickles":50, "knuts":25}
print(total(**coins), "Knuts")

#here we are using dictionary and unpacking it
```

```python
def f(*args, **kwargs):
    print("Positional:", args)

f(100,50,25)
#this f is calling function f and passing positional arguments

#    *args means: “collect all positional arguments into a tuple named args”
#   **kwargs means: “collect all keyword arguments into a dict named kwargs”
```

```python
# You receive user data as a dict (from a form / API), and you want to create an object cleanly.
class User:
    def __init__(self, name, email, age):
        self.name = name
        self.email = email
        self.age = age

payload = {"name": "Pradip", "email": "pradip@example.com", "age": 17}

user = User(**payload)   # same as User(name="Pradip", email="...", age=17)
print(user.name, user.email, user.age)
```
