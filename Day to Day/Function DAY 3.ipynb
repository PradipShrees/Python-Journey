```python
# Match-case function
name = input("Enter your name: ")

match name:
    case "Harry":
        print("Welcome to Harry")
    case "Tony":
        print("Welcome to Tony")
    case "Ron":
        print("Welcome to Ron")
    case _:
        #this _: acts as else
        print("Who?")
```

```python
# Match-case function
name = input("Enter your name: ").title().strip()

match name:
    case "Harry", "Ben" , "Tony":
        #comma will not act as or
        #if ypu want to add or use | like
        # case "Harry" | "Ben" | "Tony":
        print(f" Welcome{name}")
    case _:
        #this _: acts as else
        print("Who?")
```

```python
def main():
    num = get_number()
    meow(num)

def get_number():
    while True:
        n = int(input("Enter a number:"))
        if n > 0:
            return n

def meow(n):
    for _ in range(n):
            print("meow")
main()
```

```python
# Dict -Dictionary
# loop and list
students = {"Ben": "Town1",
       "Gokul": "Town2",
       "Tony": "Town3",
       "Bijina": "Town4"

       }
#Key/Value  rule
for student in students:
    print(student, students[student], sep="---->")
    # Here student means k
    """	sep = between values in one print call
	end = what to put at the end of that print line (default is newline \n)"""

#print(students["Bijina"])
```

```python
# Dict -Dictionary
# loop and list
students = [{"name" : "Ben", "house" : "room 1", "Zip":"2216"},
            { "name" : "Bijina", "house" : "room 2", "Zip":"2217"},
            { "name" : "Jimin", "house" : "room 3", "Zip":"2218" },
            { "name" : "Tony", "house" : "room 4", "Zip":None}
            ]
#Key/Value  rule
for student in students:
    print(student["name"], student["house"], student["Zip"], sep="----->")
```
